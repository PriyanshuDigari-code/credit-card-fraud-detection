```python
import numpy as numpy
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
credit_card_data = pd.read_csv("/content/creditcard.csv")
```

```python
credit_card_data.head(10)
```

```python
credit_card_data.tail(10)
```

```python
credit_card_data.info()
```

```python
credit_card_data.isnull().sum()
```

```python
credit_card_data["Class"].value_counts()
```

```python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.head(5)
```

```python
fraud.head(5)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
credit_card_data.groupby("Class").mean()
```

```python
legit_sample = legit.sample(n=492)
```

```python
new_dataset = pd.concat([legit_sample,fraud], axis=0)
```

```python
new_dataset.head(5)
```

```python
new_dataset.shape
```

```python
new_dataset["Class"].value_counts()
```

```python
new_dataset.groupby("Class").mean()
```

```python
X = new_dataset.drop(columns="Class",axis=1)
Y = new_dataset["Class"]
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on Training Data : ", training_data_accuracy)
```

```python
training_data_f1_score = f1_score(X_train_prediction,Y_train)
```

```python
print("F1 Score on Training Data : ", training_data_f1_score)
```

```python
training_data_confusion_matrix = confusion_matrix(X_train_prediction,Y_train)
```

```python
print("Confusion Matrix on Training Data : ", training_data_confusion_matrix)
```

```python
training_data_classification_report = classification_report(X_train_prediction,Y_train)
```

```python
print("Classification Report on Training Data : ", training_data_classification_report)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy on Test Data : ", test_data_accuracy)
```

```python
test_data_f1_score = f1_score(X_test_prediction,Y_test)
```

```python
print("F1 Score on Test Data : ", test_data_f1_score)
```

```python
test_data_confusion_matrix = confusion_matrix(X_test_prediction,Y_test)
```

```python
print("Confusion Matrix on Test Data : ", test_data_confusion_matrix)
```

```python
test_data_classification_report = classification_report(X_test_prediction,Y_test)
```

```python
print("Classification Report on Test Data : ", test_data_classification_report)
```
